```python
it_count=0
def call_player_stats(x):
    global it_count
    a = get_player_stats_last_five_games(int(x.player_id), int(x.timestamp))
    x = x.drop(index=['player_id'])
    series = pd.concat([a, x], axis=0)
    it_count = it_count+1
    print(it_count)
    return series

df = pd.DataFrame()
df = df.append((defenders.apply(call_player_stats, axis=1)), ignore_index=True)
```

```python
df = df.fillna(0)
df = df.astype({"player_id": int, "fixture_id": int, "timestamp": int})
display(df)
```
