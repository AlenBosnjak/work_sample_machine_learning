```python
import sqlite3
import pandas as pd
```

```python
path = r"XXXX-XXXX"
con = sqlite3.connect(path)
cur = con.cursor()
forwards = pd.read_sql_query("SELECT *  FROM forwardRating", con)
```

```python
forwards.drop(columns='index', inplace=True)
forwards = forwards.sort_values(by=['f_rating'], ascending=False, ignore_index=True)
display(forwards)
```

```python
test = pd.read_sql_query("SELECT DISTINCT player_id FROM forwardRating WHERE player_id IN (SELECT player_id FROM forwardRating ORDER BY f_rating DESC LIMIT 200);", con)
display(test)
```

```python
for i in range(len(test)):
    player_id = str(test["player_id"][i])
    name = pd.read_sql_query("SELECT  * FROM lineups WHERE player_id = (?) AND player_position = 'F';", con, params=(player_id,))
    print(str(i) + ". " + name['player_name'][0])
```
