```python
import sqlite3
import pandas as pd
```

```python
path = r"XXXX-XXXX"
con = sqlite3.connect(path)
cur = con.cursor()
teamratings = pd.read_sql_query("SELECT *  FROM teamRatings", con)
```

```python
teamratings.drop(columns='index', inplace=True)
teamratings.describe()
```

```python
teamratings.head()
```

```python
matches = pd.read_sql_query("SELECT * FROM fixtures WHERE season = 2020 AND league_id = 78 AND (home_id = 157 OR away_id = 157)", con)
display(matches)
```

```python
dataset = pd.merge(teamratings, matches, how='inner', on=['fixture_id'])
display(dataset)
```

```python
#Spiele des letzten Spieltages
matches_all = pd.read_sql_query("SELECT * FROM fixtures WHERE season = 2020 AND league_id = 78 AND league_round = 'Regular Season - 34'", con)
display(matches_all)
```

```python
#TeamRatings nach dem letzten Spieltag
dataset = pd.merge(teamratings, matches_all, how='inner', on=['fixture_id'])
display(dataset)
```

```python
#sortiert nach elos
elos = dataset[["team_id", "elo", "attack", "defence", "home_id", "home_name", "away_id", "away_name" ]]
elos = elos.sort_values(by=['elo'], ascending=False, ignore_index=True)
display(elos)
```

```python
#Mannschaften sortiert nach elo am ende der Saison
for i in range(len(elos)):
    if(elos['team_id'][i] == elos['home_id'][i]): 
        print(elos['home_name'][i] + ": "+ str(elos['elo'][i]))
    else:
        print(elos['away_name'][i] + ": "+ str(elos['elo'][i]))
```
